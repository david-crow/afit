```python
from statsmodels.tsa.stattools import grangercausalitytests as granger
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import pyEDM as edm
import numpy as np
import time
import os
```

```python
# vehicle specifications
wheel_diameter = 0.7620 # 30 inches
wheelbase = 1.8288 # 72 inches
track = 1.5240 # 60 inches
speed = 11.176 # 25 mph
rpm = 60 * speed / wheel_diameter / np.pi # v = (rpm) * d * pi / 60

# steering specifications
max_turning = 45
max_steering = 360
steering_ratio = max_steering / max_turning

# other parameters
num_observations = 1000
num_plotted = num_observations
```

```python
def compute_turning(steering_angles):
    return steering_angles / steering_ratio

def compute_ratios(turning):
    mask = turning != 0
    
    inside_turning = np.deg2rad(np.abs(turning))
    inside_horizontal = np.ones_like(turning)
    inside_radius = np.ones_like(turning)
    outside_turning = np.ones_like(turning)
    outside_horizontal = np.ones_like(turning)
    outside_radius = np.ones_like(turning)
    
    inside_horizontal[mask] = wheelbase / np.tan(np.abs(inside_turning[mask]))
    inside_radius[mask] = wheelbase / np.sin(np.abs(inside_turning[mask]))

    outside_horizontal[mask] = inside_horizontal[mask] + track
    outside_turning[mask] = np.arctan(wheelbase / outside_horizontal[mask])
    outside_radius[mask] = wheelbase / np.sin(outside_turning[mask])
    
    middle_radius = (inside_radius + outside_radius) / 2
    inside_radius[mask] /= middle_radius[mask]
    outside_radius[mask] /= middle_radius[mask]
    
    return np.where(turning >= 0, (outside_radius, inside_radius), (inside_radius, outside_radius)).T

def compute_rpms(ratios):
    return ratios.T[0] * rpm, ratios.T[1] * rpm
```

```python
# generate x and y
times = np.arange(0, num_observations, 1)
steering = (
    90 * np.sin(0.01 * times) +
    90 * np.sin(0.02 * times) +
    30 * np.sin(0.03 * times) + 
    30 * np.sin(0.04 * times) +
    10 * np.sin(0.05 * times) +
    45 * np.sin(0.04 * times) +
    45 * np.sin(0.05 * times) +
    15 * np.sin(0.05 * times)
)

# scale and clip y
steering = steering * max_steering / np.abs(min(np.min(steering), np.max(steering)))
steering = np.clip(steering, -max_steering, max_steering)

# plot it
fig = plt.figure(figsize=(20,5))
plt.plot(times[:num_plotted], steering[:num_plotted])
plt.plot([0, num_plotted], [max_steering, max_steering], linestyle="--", color="k")
plt.plot([0, num_plotted], [-max_steering, -max_steering], linestyle="--", color="k")
plt.plot([0, num_plotted], [0, 0], linestyle="--", color="k")
```

```python
# compute rpms
left_rpms, right_rpms = compute_rpms(compute_ratios(compute_turning(steering)))

# save these
labels = ["Steering wheel angle", "Left wheel RPM", "Right wheel RPM"]
label_size = 16
tick_size = 12

# save to a file
df = pd.DataFrame(np.vstack((
    np.array(steering),
    np.array(left_rpms),
    np.array(right_rpms))))
df = df.T
df.columns = labels
df.to_csv("/Users/david/Documents/masters/thesis/data/steering/steering.csv", index_label="Time")

# plot everything
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(times, steering, label=labels[0], linewidth=3, linestyle="-")
ax.plot(times, left_rpms, label=labels[1], linewidth=3, linestyle="-")
ax.plot(times, right_rpms, label=labels[2], linewidth=3, linestyle="-")

ax.set_xlabel("Time (seconds)", fontsize=label_size)
ax.set_ylabel("Value", fontsize=label_size)
ax.xaxis.set_tick_params(labelsize=tick_size)
ax.yaxis.set_tick_params(labelsize=tick_size)

ax.legend(prop={"size": tick_size})
fig.tight_layout()
# fig.savefig("/Users/david/Documents/masters/thesis/images/SteeringSystemTimeSeries.png", dpi=200)
```

```python
# normalize and save
steering = ((steering - np.min(steering)) / np.ptp(steering)) * 2 - 1
left_rpms = ((left_rpms - np.min(left_rpms)) / np.ptp(left_rpms))
right_rpms = ((right_rpms - np.min(right_rpms)) / np.ptp(right_rpms))

# plot everything
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(times, steering, label=labels[0], linewidth=3, linestyle="-")
ax.plot(times, left_rpms, label=labels[1], linewidth=3, linestyle="-")
ax.plot(times, right_rpms, label=labels[2], linewidth=3, linestyle="-")

ax.set_xlabel("Time to Prediction", fontsize=label_size)
ax.set_ylabel("Value", fontsize=label_size)
ax.xaxis.set_tick_params(labelsize=tick_size)
ax.yaxis.set_tick_params(labelsize=tick_size)

ax.legend(prop={"size": tick_size})
fig.tight_layout()
# fig.savefig("/Users/david/Documents/masters/thesis/images/SteeringSystemTimeSeriesScaled.png", dpi=200)
```

```python
# test for granger causality
for i, series1 in enumerate([steering, left_rpms, right_rpms]):
    for j, series2 in enumerate([steering, left_rpms, right_rpms]):
        if i != j:
            print("{}\nDoes {} Granger cause {}?\n".format("-" * 20, labels[j].lower(), labels[i].lower()))
            
            plt.plot(times, series1, label=labels[i])
            plt.plot(times, series2, label=labels[j])
            plt.legend()
            plt.show()
            
            test_vals = np.vstack((series1, series2)).T
            results = granger(test_vals, maxlag=5)
            p_val = results[1][0]["params_ftest"][1]
            
            # check null hypothesis
            if p_val < 0.05:
                print("\n{} g-causes {} (p = {:.4f})\n".format(labels[j].lower(), labels[i].lower(), p_val))
            else:
                print("\n{} doesn't g-cause {} (p = {:.4f})\n".format(labels[j].lower(), labels[i].lower(), p_val))
```

```python
# test for granger causality
labels = ["Steering wheel angle", "Left wheel RPM", "Right wheel RPM"]
maxlags = 20

xy_labels = []
lags = []
p_vals = []

for i, series_y in enumerate([steering, left_rpms, right_rpms]):
    for j, series_x in enumerate([steering, left_rpms, right_rpms]):
        if i != j:
            test_vals = np.vstack((series_y, series_x)).T
            results = granger(test_vals, maxlag=maxlags, verbose=False)
            
            for k in range(1, maxlags + 1):
                xy_labels.append("{} & {}".format(labels[j], labels[i]))
                lags.append(k)
                
                p_vals.append(np.average([
                    results[k][0]["ssr_ftest"][1],
                    results[k][0]["ssr_chi2test"][1],
                    results[k][0]["lrtest"][1],
                    results[k][0]["params_ftest"][1]
                ]))

gdf = pd.DataFrame(np.hstack((
    np.array(xy_labels).reshape(-1,1),
    np.array(lags).reshape(-1,1),
    np.array(p_vals).reshape(-1,1)
)), columns=["Label", "Lags", "P"])

gdf.to_csv("/Users/david/Documents/masters/thesis/data/steering/steering-granger.csv", index=False)
```
