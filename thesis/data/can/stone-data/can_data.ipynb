```python
import pandas as pd
import numpy as np

import time
import sys
import os
```

```python
metadata = pd.read_csv("metadata.csv")
display(metadata)
```

```python
# how long does this take?
start = time.time()

# make the output file
output = open("can_data.csv", "w+")
output.write("captureid,vehicle_id,ts,arbitration_id,dlc,data,make,model,year\n")

# for every vehicle...
for i in metadata["Capture"].unique():
    vehicle_start = time.time()
    print("Starting vehicle {}...".format(i), end="\t")
    
    f = open("temp/loggerProgram{}.log".format(i), "r")
    f.readline() # ignore the header

    # for every message...
    for line in f:
        # remove white space and the timestamp's ":"
        line = line.split()
        line[0] = line[0][:-1]
        
        # because sometimes pandas indices don't line up
        meta = metadata[metadata["Capture"] == i]
        
        # build the output string
        out = [
            str(i),                     # captureid
            str(i),                     # vehicle_id
            str(line[0]),               # timestamp3
            str(line[1]).upper(),       # arbID
            str(line[2]),               # dlc
            "".join(line[3:]).upper(),  # data
            str(meta["Make"].iloc[0]),  # make
            str(meta["Model"].iloc[0]), # model
            str(meta["Year"].iloc[0])   # year
        ]
        
        # write to the file
        output.write(",".join(out) + "\n")

    f.close()
    vehicle_elapsed = time.time() - vehicle_start
    print("Done! Elapsed: {:.2f} seconds ({:.2f} minutes)".format(vehicle_elapsed, vehicle_elapsed / 60))

# output.close()
elapsed = time.time() - start
print("Elapsed time: {:.2f} seconds ({:.2f} minutes)".format(elapsed, elapsed / 60))
```
