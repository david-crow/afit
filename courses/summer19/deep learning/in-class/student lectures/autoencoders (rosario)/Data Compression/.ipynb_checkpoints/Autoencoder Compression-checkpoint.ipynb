```python
'''Taken from Neural Netowrk Projects with Python
Section - Removing Noise from Images Using Autoencodes
    subsection - Building a simple autoencoder
        subsubsection - Effectf of hidden layer size on autoencoders performance
'''
from keras.models    import Sequential, load_model
from keras.layers    import Dense
from keras.datasets  import mnist
from IPython.display import Image
from keras.utils import np_utils, plot_model
from keras import callbacks
from IPython.display import Image

from matplotlib import pyplot as plt
#make plots inline using jupyter magic
%matplotlib inline

import numpy as np
import random, warnings, os
warnings.filterwarnings('ignore')
```

```python
def graphHistory(history, title='',metrics=['loss'], fontsize=20,
                linewidth=5,figsize=(10,10), tick_size = 5):
    '''Function for graphing the training and valiedation accuracy and loss'''
    
    plt.rc('xtick', labelsize=tick_size) 
    plt.rc('ytick', labelsize=tick_size) 
        
    # Plot history for different metrics 
    for strText in metrics:
        fig = plt.figure()
        plt.figure(figsize = figsize)
        plt.plot(history.history[strText], linewidth=linewidth)
        
        try:
            plt.plot(history.history['val_'+strText],linewidth=linewidth )
            plt.legend(['Tr. '+strText, 'Val. '+strText], 
                       fontsize =fontsize)
        except:
            plt.legend(['Tr. '+strText],fontsize =fontsize)
                        
        plt.title('MODEL '+strText.upper() + title,fontsize =fontsize)
        plt.ylabel(strText.upper(),fontsize=fontsize)
        plt.xlabel('Epoch',fontsize=fontsize)           
        plt.show()

def save_model(model, save_dir=os.path.join(os.getcwd(), 'saved_models'),
                   model_file_name='model.h5'):
    """
    Save model and current weights
    :param model: Keras model
    :param save_dir: path name to save directory
    :param model_file_name: filename for saved model
    :return: nothing
    """
    if not os.path.isdir(save_dir):
        os.makedirs(save_dir)
    model_path = os.path.join(save_dir, model_file_name)
    model.save(model_path)
    print('Saved trained model at %s ' % model_path)
    
    
def loadmodel(model_file_name, directory='saved_models'):
    # Load model and weights
    if not os.path.isdir(directory):
        os.makedirs(directory)
    model_path = os.path.join(directory, model_file_name)
    model = load_model(model_path)
    print('Loaded trained model from %s ' % model_path)
    return model         
```

# Step 1 Data preprocessing 

```python
training_set, testing_set = mnist.load_data()
X_train, y_train = training_set
X_test, y_test = testing_set
print(X_train.shape, X_test.shape)
```

Visualize the data

```python
fig, ((ax1,ax2,ax3,ax4,ax5),(ax6,ax7,ax8,ax9,ax10)) = plt.subplots(2,5,figsize=(10,5))
for idx, ax in enumerate([ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9,ax10]):
    for i in range(1000):
        if y_test[i] == idx:
            ax.imshow(X_test[i],cmap='gray')
            ax.grid(False)
            ax.set_xticks([])
            ax.set_yticks([])
            break
plt.tight_layout()
plt.show()
```

## Reshape the images from 28 x 28 to 784 x 1 and Normalize the values for 0  to 255 to a scale 0-1

```python
X_trained_reshaped = X_train.reshape((X_train.shape[0],
                                      X_train.shape[1]*X_train.shape[2]))/255.
X_test_reshaped = X_test.reshape((X_test.shape[0],
                                      X_test.shape[1]*X_test.shape[2]))/255.
```

The the number of nodes in the hidden layer is a hyperparameter and it must be selected to ensure that the latent representation is a compressed representation which enables the reconstruction of the original input.

# Step 2 Coding the Autoencoder Change image 

## Have a function to creat models, in order to optimize the hidden_layer_size hyper-parameter

```python
def create_basic_autoencoder(hidden_layer_size = 1, 
                             input_shape=(784,),   
                             activation = 'relu'):
    model = Sequential()    
    model.add(Dense(units = hidden_layer_size, 
                    input_shape = input_shape,
                    activation = 'relu')) 

    #We use a sigmoid activation function to constraint the output values from 0 to 1
    model.add(Dense(units = input_shape[0], activation = 'sigmoid'))
    return model
```

```python
model = create_basic_autoencoder()

#The mean_squared_error is useful in this case because we need a loss function that quantifies 
#the pixel-wise discrepancy between the input and the output
model = create_basic_autoencoder(hidden_layer_size = 1)
model.compile(optimizer = 'adam',loss='mean_squared_error')

plot_model(model, show_shapes=True,to_file='autoencoder1.png')
Image(filename='autoencoder1.png') 
```

# Fun facts: Becuase we are interested in autoencoder producing a output identical to the input, we pass the model the same data for both X and Y.

# Step 3 Training the model

```python
%%time
#Now we train
history = model.fit(X_trained_reshaped, 
                    X_trained_reshaped, 
                    epochs = 20, 
                    verbose=0, 
                    batch_size=512)
save_model(model,model_file_name='model1.h5' )
graphHistory(history, title='',metrics=['loss'],linewidth=4,figsize=(15,5),
                    tick_size=20,fontsize=20)
```

## Apply the test data to the trained model


```python
output = model.predict(X_test_reshaped)
```

```python
fig, ((ax1,ax2,ax3,ax4,ax5),
      (ax6,ax7,ax8,ax9,ax10)) = plt.subplots(2,5,figsize=(20,7))

#Randomly select 5 number to display the results
rand_selection = random.sample(range(output.shape[0]),5)

#Plot original images (input) on top row
for idx, ax in enumerate([ax1,ax2,ax3,ax4,ax5]):
    ax.imshow(X_test[rand_selection[idx]],cmap='gray')
    
    if idx == 0:
        ax.set_ylabel("INPUT", size=40)
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
        
#Plot output images on the button row
for idx, ax in enumerate([ax6,ax7,ax8,ax9,ax10]):
    ax.imshow(output[rand_selection[idx]].reshape(28,28),cmap='gray')
    
    if idx == 0:
        ax.set_ylabel("OUTPUT", size=40)
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    
            
plt.tight_layout()
plt.show()
```

## Train 5 models with hidden layer size of 2,4,8,16 and 32 neurons. Predict on the test data and store the output for plotting

```python
%%time
print('Training began...')
model2 = create_basic_autoencoder(hidden_layer_size = 2)
model2.compile(optimizer = 'adam',loss='mean_squared_error')
model2.fit(X_trained_reshaped, X_trained_reshaped, epochs = 10,verbose = 0)
save_model(model2, model_file_name='model2.h5')

model4 = create_basic_autoencoder(hidden_layer_size = 4)
model4.compile(optimizer = 'adam',loss='mean_squared_error')
model4.fit(X_trained_reshaped, X_trained_reshaped, epochs = 10,verbose = 0)
save_model(model4, model_file_name='model4.h5')

model8 = create_basic_autoencoder(hidden_layer_size = 8)
model8.compile(optimizer = 'adam',loss='mean_squared_error')
model8.fit(X_trained_reshaped, X_trained_reshaped, epochs = 10,verbose = 0)
save_model(model8, model_file_name='model8.h5')

model16 = create_basic_autoencoder(hidden_layer_size = 16)
model16.compile(optimizer = 'adam',loss='mean_squared_error')
model16.fit(X_trained_reshaped, X_trained_reshaped, epochs = 10,verbose = 0)
save_model(model16, model_file_name='model16.h5')

model32 = create_basic_autoencoder(hidden_layer_size = 32)
model32.compile(optimizer = 'adam',loss='mean_squared_error')
model32.fit(X_trained_reshaped, X_trained_reshaped, epochs = 10, verbose = 0)
save_model(model32, model_file_name='model32.h5')

model64 = create_basic_autoencoder(hidden_layer_size = 64)
model64.compile(optimizer = 'adam',loss='mean_squared_error')
model64.fit(X_trained_reshaped, X_trained_reshaped, epochs = 10, verbose = 0)
save_model(model64, model_file_name='model64.h5')

print('Training completed.')
```

# Step 4 Plot the results

```python
#Load the models
model1  = loadmodel('model1.h5')
model2  = loadmodel('model2.h5')
model4  = loadmodel('model4.h5')
model8  = loadmodel('model8.h5')
model16 = loadmodel('model16.h5')
model32 = loadmodel('model32.h5')
model64 = loadmodel('model64.h5')

output   = model1.predict(X_test_reshaped)
output2  = model2.predict(X_test_reshaped)
output4  = model4.predict(X_test_reshaped)
output8  = model8.predict(X_test_reshaped)
output16 = model16.predict(X_test_reshaped)
output32 = model32.predict(X_test_reshaped)
output64 = model64.predict(X_test_reshaped)
```

```python
fig, axes = plt.subplots(8,5,figsize=(32,32))

#Randomly select 5 number to display the results
rand_selection = random.sample(range(output.shape[0]),5)
outputs = [X_test, output, output2,output4,output8,output16,output32,output64]

layerSizes = [0,1,2,4,8,16,32,64]
#Plot original images (input) on top row
for row_num, row in enumerate(axes):
    for col_num, ax in enumerate(row):
        ax.imshow(outputs[row_num][rand_selection[col_num]].reshape(28,28),
                  cmap='gray')
        if row_num == 0 and col_num == 0:
            ax.set_ylabel("Input\n Images", fontsize=40)
        elif col_num == 0:
            ax.set_ylabel("{} Nodes".format(
                layerSizes[row_num]), fontsize=40)

        ax.grid(False)
        ax.set_xticks([])
        ax.set_yticks([])      
            
plt.tight_layout()
plt.show()
```

At 32 nodes in the hidden layer, the output becomes very close to the original input. That is 24.5 compression ratio (784/32) and we still obtain a satisfactory output. Remember that compression factor can be a very important parameter to consider, as mapping the input to a very low dimensional space will result in too much information loss, leading to poor reconstructions. There may simply not be enough space to store the key essentials of the image. On the other hand, we are already aware of how providing our model with too much learning capacity may cause it to overfit, which is why choosing a compression factor by hand can be quite tricky. When in doubt, it can never hurt to experiment with different compression factors as well as regularization methods.

# Step 5 Implement Signal compression

```python
length = 784
Amplitude = range(1,10+1)
Frequency = range(500,1000, 5)
Phase = range(0,91)
t = np.linspace(0,.01,length)

rows = len(Amplitude)*len(Frequency)*len(Phase)
signals = np.zeros([rows, length])

i = 0
for a in Amplitude:
    for f in Frequency:
        for phi in Phase:
            signals[i,:] = a*np.cos(2*np.pi*f*t+phi) 
            i+=1

print(signals.shape)
plt.rc('xtick', labelsize=40)     
plt.rc('ytick', labelsize=40)

fig = plt.figure(figsize=(25,10))
plt.plot(signals[0,:], linewidth=10);
plt.plot(signals[-1,:],  linewidth=10);
```

# Normalizing and spliting the dataset

```python
Signals_normalizes = (signals+10.0)/20.0  #Offset and normlize
del signals
fig = plt.figure(figsize=(25,10))
plt.plot(Signals_normalizes[0,:], linewidth=10);
plt.plot(Signals_normalizes[-1,:], linewidth=10);
```

```python
np.random.shuffle(Signals_normalizes)
Train, Test = Signals_normalizes[0:90_000], Signals_normalizes[90_000:]
del Signals_normalizes
print(Train.shape, Test.shape)
```

# Training the models 

```python
def create_basic_autoencoder2(hidden_layer_size = 1, 
                             input_shape=(784,),
                             act_func = 'relu'):
    model = Sequential()    
    model.add(Dense(units = hidden_layer_size, 
                    input_shape = input_shape,
                    activation = act_func)) 

    #----------------------- Student ----------------------------
    #             Change n to the correct output size 
    #------------------------------------------------------------
    output = 0
    
    
    
    model.add(Dense(units = output, activation = 'sigmoid'))
    return model
```

```python
%%time
print('Training began...')


model8 = create_basic_autoencoder2(hidden_layer_size = 8,
                                   input_shape=(784,))
model16 = create_basic_autoencoder2(hidden_layer_size = 16,
                                   input_shape=(784,))
model32 = create_basic_autoencoder2(hidden_layer_size = 32,
                                   input_shape=(784,))


model8.compile(optimizer = 'adam',loss='mean_squared_error')
model16.compile(optimizer = 'adam',loss='mean_squared_error')
model32.compile(optimizer = 'adam',loss='mean_squared_error')


#----------------------- Student ----------------------------
#   Add the correct labels to train the models
#------------------------------------------------------------

history = model8.fit(Train,  'HERE', epochs = 10, verbose = 0)
history = model16.fit(Train, 'HERE', epochs = 10, verbose = 0)
history = model32.fit(Train, 'HERE', epochs = 10, verbose = 0)


save_model(model8, model_file_name='model8s.h5')
graphHistory(history, title='',metrics=['loss'],linewidth=4,figsize=(15,5),
                    tick_size=20,fontsize=20)
save_model(model16, model_file_name='model16s.h5')
graphHistory(history, title='',metrics=['loss'],linewidth=4,figsize=(15,5),
                    tick_size=20,fontsize=20)

save_model(model32, model_file_name='model32s.h5')
graphHistory(history, title='',metrics=['loss'],linewidth=4,figsize=(15,5),
                    tick_size=20,fontsize=20)

print('Training completed.')
```
