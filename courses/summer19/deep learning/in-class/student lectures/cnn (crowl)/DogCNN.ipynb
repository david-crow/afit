# CNN definitions

### makeCNN() is my own and makeCNN2() comes from the internet.  These are both relatively simple because 1) it's a lot of work to get good accuracy on this kind of problem and 2) we'll see later that a good classifier is so much bigger

```python
def makeCNN():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', data_format="channels_last", input_shape=(IMG_SIZE,IMG_SIZE,3)))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(0.01)))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(256, (3,3), activation='relu'))
    model.add(MaxPooling2D((2,2)))
    model.add(Flatten())
    model.add(Dense(512, activation='tanh'))
    model.add(Dense(8, activation='softmax', kernel_regularizer=regularizers.l2(0.01)))
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
'''
https://towardsdatascience.com/dog-breed-classification-using-cnns-f042fbe0f333
Dog Breed Classification Using CNNs
Deniz Doruk Nuhoglu
10 Feb
'''
def makeCNN2():
    # CNN model
    model = Sequential()

    # -----------------------------------------------------------------------------------
    # conv 1
    model.add(Conv2D(16, (3,3), input_shape=(IMG_SIZE,IMG_SIZE,3)))
    model.add(BatchNormalization(axis=3))
    model.add(Activation('relu'))
    #model.add(Dropout(0.5))

    # max pool 1
    model.add(MaxPooling2D(pool_size=(2,2),strides=2))                             

    # -----------------------------------------------------------------------------------
    # # conv 2
    model.add(Conv2D(32, (3,3)))                                                   
    model.add(BatchNormalization(axis=3))
    model.add(Activation('relu'))
    #model.add(Dropout(0.5))

    # max pool 2
    model.add(MaxPooling2D(pool_size=(2,2),strides=2))                             
    # -----------------------------------------------------------------------------------

    # conv 3
    model.add(Conv2D(48, (3,3)))                                                   
    model.add(BatchNormalization(axis=3))
    model.add(Activation('relu'))
    #model.add(Dropout(0.7))

    # max pool 3
    model.add(MaxPooling2D(pool_size=(2,2),strides=2))                             
    # -----------------------------------------------------------------------------------

    # # conv 4
    model.add(Conv2D(64, (3,3)))                                                   
    model.add(BatchNormalization(axis=3))
    model.add(Activation('relu'))
    #model.add(Dropout(0.7))
    # max pool 4
    model.add(MaxPooling2D(pool_size=(2,2),strides=2))                             

    # flatten
    model.add(Flatten())                                                           

    # fc layer 1
    model.add(Dense(1024, activation='relu'))                                  

    # fc layer 2
    model.add(Dense(512, activation='relu'))

    # fc layer 3
    model.add(Dense(256, activation='relu'))

    # fc layer 4
    model.add(Dense(2, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

# Generators for training and validation sets

This is the crux of today's lesson.  Python generators may create infinite data if asked to do so, which they will do via bootstrapping.  They also allow for easy incorporation of data augmentation techniques, as seen in the arguments to the function.

```python
def getGenerators(rotRange=0, shift=0.0, flipVert=False, flipHorz=False):

    ##########################################################################################################
    # Student code here
      # Create an ImageDataGenerator that applies the in-passed augmentation techniques
      # Use a validation split of .2
      # don't forget to rescale the pixels to be in a range from 0-1
    ##########################################################################################################
    train_datagen = ImageDataGenerator(
        rotation_range = rotRange,
        horizontal_flip = flipHorz,
        vertical_flip = flipVert,
        width_shift_range = shift,
        height_shift_range = shift,
        rescale = 1. / 255,
        validation_split = 0.2)
    
    # You may have to adjust the batch size depending on the memory available to you
    train_gen = train_datagen.flow_from_directory(
        imageDirectory,
        target_size=(IMG_SIZE,IMG_SIZE),
        batch_size=20,
        class_mode='categorical',
        subset='training')

    val_gen = train_datagen.flow_from_directory(
        imageDirectory,
        target_size=(IMG_SIZE,IMG_SIZE),
        batch_size=27, # 27 images per batch, 10 steps per epoch = 270 images.  Perfect!
        class_mode='categorical',
        subset='validation')
    
    return train_gen, val_gen
```

## Make the model and view the summary

```python
dogCNN = makeCNN()
dogCNN.summary()
```

## Create a training and validation set with no data augmentation

```python
train_gen, val_gen = getGenerators()
```

## Use fit_generator() just like fit() to conduct training and validation

```python
start = time.time()


history = dogCNN.fit_generator(train_gen, steps_per_epoch=55, epochs=15, verbose=1,
                               validation_data=val_gen, validation_steps=10)#,
                               #callbacks=[ModelCheckpoint("./model.h5", save_best_only=True, monitor="val_acc")])

end = time.time()
print("Run time: ", '%.2f'%((end - start) / 60), " minutes")
```

#### About 25% validation accuracy isn't great, but it will suffice for this simple walkthrough

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

Notice that the training accuracy continued to increase.  Because we have such a small dataset (1300 pictures), the CNN quickly memorized the pictures.  Let's try again with data augmentation.  By specifying arguments to getGenerators(), the training_generator and validation_generator will automatically and randomly rotate, shift, and flip the images as they get fed into the CNN.  This will help keep the CNN on its toes, so to speak.

```python
# Used False for vertical flip, because let's keep it kind of simple for now
train_gen_with_augmentation, val_gen_with_augmentation = getGenerators(5, 0.2, False, True)
```

```python
dogCNN_with_Augmentation = makeCNN()
```

Let's do 30 epochs this time to ensure the data augmentation works as expected.

```python
start = time.time()

if(os.path.exists("./DogClassifierWithAugmentation.h5")):
    dogCNN_with_Augmentation = load_model("DogClassifierWithAugmentation.h5")
else:
    history_with_augmentation = dogCNN_with_Augmentation.fit_generator(train_gen_with_augmentation, steps_per_epoch=54, epochs=30, verbose=1,
                               validation_data=val_gen_with_augmentation, validation_steps=10)#,
                               #callbacks=[ModelCheckpoint("./model.h5", save_best_only=True, monitor="val_acc")])
    
    save_model(dogCNN_with_Augmentation, "DogClassifierWithAugmentation.h5")  

end = time.time()
print("Run time: ", '%.2f'%((end - start) / 60), " minutes")
```

#### Now we're up to 35% accuracy!  Still not amazing, but 10% better than before, and we didn't have to change our network.  If you look at the training accuracy, you'll notice that we weren't overfitting the data, too.

```python
acc = history_with_augmentation.history['acc']
val_acc = history_with_augmentation.history['val_acc']
loss = history_with_augmentation.history['loss']
val_loss = history_with_augmentation.history['val_loss']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

# Make predictions with new doggo pictures

```python
def predictPicture(filename, top=len(breeds)):
    ##########################################################################################################
    # Student code here
    # Load the image from the in-passed filename
    # Don't forget to specify the size of the image using IMG_SIZE
    ##########################################################################################################
    img = load_img(testDirectory + filename, target_size=(IMG_SIZE, IMG_SIZE))
    
    ##########################################################################################################
    # Turn the image into a format (tensor) the CNN can use
    # Hint: use img_to_array()
    ##########################################################################################################
    img_tensor = img_to_array(img)
    print(img_tensor.shape)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    print(img_tensor.shape)
    
    ##########################################################################################################
    # Use the .predict function like we're used to in previous machine learning tasks
    ##########################################################################################################
    y_prob = dogCNN_with_Augmentation.predict(img_tensor)
    
    # y_prob will yield an array of probabilities - not much use if we don't know which breeds correspond to which values
    # Sort them too, because that's always nice to see and we don't hate our future selves.
    prob = sorted(list(zip(breeds, y_prob[0])), reverse=True, key=lambda prob: prob[1])
    
    print(testDirectory + filename)
    print(prob[:top])
    print()
```

```python
predictPicture("german_shepherd.jpg") # Failure
predictPicture("labrador.jpg")        # Failure
predictPicture("pembroke.jpg", 3)     # Failure.  This simple classifier stinks!
predictPicture("borderCollie.jpg", 3) # Smashing success! 
                                      # The classifier must have learned well on something distinctive to a border collie
```
