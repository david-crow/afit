## Load the training/testing images

```python
def load_images(filepath, prefix):
    # we'll need this
    x = []
    
    # key: alphabet
    # value: two-element list ([global index of first image; global index of last image])
    ranges = {}
    
    # global index
    idx = 0
    
    # for the user
    print("{}: found {} alphabets\n".format(prefix, len(os.listdir(filepath))))
    
    # for every alphabet...
    for i, alphabet in enumerate(os.listdir(filepath)):
        print("Loading alphabet {}: {}".format(i + 1, alphabet))
        ranges[alphabet] = [idx, None]
        alphabet_path = os.path.join(filepath, alphabet)
        
        # for every character in the alphabet...
        for letter in os.listdir(alphabet_path):
            letter_images = []
            letter_path = os.path.join(alphabet_path, letter)
            
            # for every image of the letter...
            for sample in os.listdir(letter_path):
                image_path = os.path.join(letter_path, sample)
                image = plt.imread(image_path)
                letter_images.append(image)
            
            # save input; update dictionary
            x.append(np.stack(letter_images))            
            ranges[alphabet][1] = idx
            idx += 1

    return np.stack(x), ranges
```

```python
# how long does this take?
start_time = time.time()

train_pickle = os.path.join(data_dir, "train.pickle")
test_pickle = os.path.join(data_dir, "test.pickle")

if os.path.exists(train_pickle) and os.path.exists(test_pickle):
    # load the stuff
    print("Loading images...", end=" ")
    
    with open(os.path.join(data_dir, "train.pickle"), "rb") as f:
        train_images, train_ranges = pickle.load(f)
    with open(os.path.join(data_dir, "test.pickle"), "rb") as f:
        test_images, test_ranges = pickle.load(f)
    
    print("done loading images!")

else:
    # training/testing sets
    train_images, train_ranges = load_images(train_dir, prefix="Training")
    test_images, test_ranges = load_images(test_dir, prefix="\nTesting")
    
    # save the stuff
    print("\nSaving images for later...", end=" ")
    
    with open(os.path.join(data_dir, "train.pickle"), "wb") as f:
        pickle.dump((train_images, train_ranges), f)
    with open(os.path.join(data_dir, "test.pickle"), "wb") as f:
        pickle.dump((test_images, test_ranges), f)
    
    print("done saving images!")

print("\nElapsed time: {:.2f} seconds".format(time.time() - start_time))
```

## Build the training/testing sets

```python
# build a set of [batch_size] image pairs
# for half of the pairs, the images come from the same class
# for the other half, the images come from different classes
def make_batch(images, batch_size):
    # save these numbers
    n_classes, n_examples, height, width = images.shape
    
    # initialize two empty arrays for the input batch (one array per image in the image pair)
    x = [np.zeros((batch_size, height, width, 1)) for i in range(2)]
    
    # randomly select [batch_size] classes
    classes = np.random.choice(n_classes, size=(batch_size,), replace=False)
    
    # for each image pair...
    for i in range(batch_size):
        # select the left character
        left_class = classes[i]
        
        # select the observation of the left character
        left_idx = np.random.randint(0, n_examples)
        
        # copy the left image into the batch
        x[0][i, :, :, :] = images[left_class, left_idx].reshape(height, width, 1)
        
        # select the right character
        if i >= batch_size // 2:
            # the first half of the pairs should have the same class
            right_class = left_class
        else: 
            # the second half of the pairs should have different classes
            right_class = (left_class + np.random.randint(1, n_classes)) % n_classes
        
        # select the observation of the right character
        right_idx = np.random.randint(0, n_examples)
        
        # copy the right image into the batch
        x[1][i, :, :, :] = images[right_class, right_idx].reshape(height, width, 1)
    
    # half of the targets should be [1] to indicate "same class"
    # the other half should be [0] to indicate "different class"
    y = np.zeros((batch_size,))    
    y[batch_size // 2:] = 1
    
    return x, y
```

## Define and test one-shot learning

```python
# make a set of [n] image pairs consisting of random images of random characters
# one pair should consist of two images from the same class
# the remaining pairs should consist of images from different classes
def make_oneshot_task(n, images, ranges, language=None):
    # save these numbers
    n_classes, n_examples, height, width = images.shape
    
    # select only characters from the given language
    if language is not None:
        # get the range of indices for the given language
        low, high = ranges[language]
        
        # ensure [n] is not too small
        if n > high - low:
            raise ValueError("This language ({}) has fewer than {} letters".format(language, n))
        
        # randomly select [n] characters from the given language
        classes = np.random.choice(range(low,high), size=(n,), replace=False)

    # otherwise, pick random characters
    else:
        # randomly select [n] characters from all languages
        classes = np.random.choice(range(n_classes), size=(n,), replace=False)
    
    # we'll need two different random numbers
    test_image, support_image, = np.random.choice(n_examples, replace=False, size=(2,))
    
    # make [n] copies of a random image from a random test class (this is the anchor)
    true_class = classes[0]
    test_images = np.asarray([images[true_class, test_image, :, :]] * n).reshape(n, height, width, 1)
    
    # select [n] random indices
    indices = np.random.randint(0, n_examples, size=(n,))
    
    # select one random image from each of the [n] random classes
    support_set = images[classes, indices, :, :]
    
    # replace the test class's random image with the previously selected image
    # we do this to guarantee that the test image for the test class
    # is different from the support image for the test class
    support_set[0, :, :] = images[true_class, support_image]
    support_set = support_set.reshape(n, height, width, 1)
    
    # the output for the first image is [1] because it's a different image for the same class
    # the output for each of the remaining images is [0] because they come from different classes
    targets = np.zeros((n,))
    targets[0] = 1
    
    # shuffle the images
    targets, test_images, support_set = shuffle(targets, test_images, support_set)
    
    # we want a set of image pairs, not two sets of images
    pairs = [test_images, support_set]

    # return the stuff
    return pairs, targets
```

```python
# evaluate [k] instances of [n]-way one-shot learning
def test_oneshot_model(model, n, k, images, ranges):
    print("Evaluating model on {} random {}-way one-shot learning tasks...".format(k,n))
    
    # count the number of correct predictions
    n_correct = 0
    
    # for each task...
    for i in range(k):
        # generate test set and predict outputs
        inputs, targets = make_oneshot_task(n, images, ranges)
        preds = model.predict(inputs)
        
        # check prediction
        if np.argmax(preds) == np.argmax(targets):
            n_correct += 1
    
    # output stats
    percent_correct = 100. * n_correct / k
    print("Average {}-way one-shot learning accuracy: {:.2f}%".format(n, percent_correct))
    return percent_correct
```

## Build the siamese neural network

See http://www.cs.utoronto.ca/~gkoch/files/msc-thesis.pdf for details on
- `initialize_weights()`,
- `initialize_bias()`, and
- `build_siamese_model()`

```python
def initialize_weights(size):
    return np.random.normal(loc=0.0, scale=1e-2, size=size)
```

```python
def initialize_bias(size):
    return np.random.normal(loc=0.5, scale=1e-2, size=size)
```

```python
def build_siamese_model(input_shape):
    # siamese networks have two inputs
    left_input = Input(shape=input_shape)
    right_input = Input(shape=input_shape)
    
    # build the model
    model = Sequential()
    model.add(Conv2D(64, (10,10), activation="relu", input_shape=input_shape,
                     kernel_initializer=initialize_weights, kernel_regularizer=l2(2e-4)))
    model.add(MaxPooling2D())
    model.add(Conv2D(128, (7,7), activation="relu", kernel_initializer=initialize_weights,
                     bias_initializer=initialize_bias, kernel_regularizer=l2(2e-4)))
    model.add(MaxPooling2D())
    model.add(Conv2D(128, (4,4), activation="relu", kernel_initializer=initialize_weights,
                     bias_initializer=initialize_bias, kernel_regularizer=l2(2e-4)))
    model.add(MaxPooling2D())
    model.add(Conv2D(256, (4,4), activation="relu", kernel_initializer=initialize_weights,
                     bias_initializer=initialize_bias, kernel_regularizer=l2(2e-4)))
    model.add(Flatten())
    model.add(Dense(4096, activation="sigmoid", kernel_initializer=initialize_weights,
                    bias_initializer=initialize_bias, kernel_regularizer=l2(1e-3)))
    
    # display the model so far
    filename = os.path.join(os.getcwd(), "images", "convolutional-model.png")
    plot_model(model, show_shapes=True, to_file=filename)
    display(Image(filename=filename, retina=True))

    # generate the encodings for the two images
    left_encoding = model(left_input)
    right_encoding = model(right_input)
    
    # add a custom layer to compute the absolute difference between the encodings
    l1_layer = Lambda(lambda tensors: K.abs(tensors[0] - tensors[1]))
    l1_distance = l1_layer([left_encoding, right_encoding])
    
    # add a dense layer to compute the similarity score
    prediction = Dense(1, activation="sigmoid", bias_initializer=initialize_bias)(l1_distance)
    
    # connect the inputs to the outputs
    siamese_net = Model(inputs=[left_input, right_input], outputs=prediction)
    
    # compile the model
    siamese_net.compile(loss="binary_crossentropy", optimizer=Adam(lr=6e-5))
    
    # return the model
    return siamese_net
```

```python
model = build_siamese_model(input_shape=(105,105,1))
```

```python
model.summary()
filename = os.path.join(os.getcwd(), "images", "siamese-model.png")
plot_model(model, show_shapes=True, to_file=filename)
display(Image(filename=filename, retina=True))
```

## Train and test the siamese neural network

```python
# this again?
def graph_history(loss_history, acc_history, n_iterations, interval):
    # create the figure
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    
    # plot the stuff
    iterations = interval * (np.arange(n_iterations // interval) + 1)
    ax1.plot(iterations, loss_history, label="Loss", color="c")
    ax2.plot(iterations, acc_history, label="Val. Accuracy", color="m")
    
    # also, indicate the iteration with the best loss
    ax1.axvline(interval * (np.argmin(loss_history) + 1), color="k", linestyle="--")

    # pretty-fy the plot
    ax1.set_xlabel("Training Iteration")
    ax1.set_ylabel("Loss")
    ax2.set_ylabel("Accuracy")
    fig.legend()

    plt.savefig(os.path.join(os.getcwd(), "images", "history.png"), dpi=100)
    plt.show()
```

```python
# hyperparameters
interval = 200  # interval for evaluating on one-shot tasks
batch_size = 32  # number of image pairs in each batch
n_iterations = 20000  # number of training iterations
n_classes = 20  # number of classes for each one-shot task
n_tasks = 250  # number of one-shot tasks to use for validation
best_acc = -1  # best validation accuracy during training
```

```python
loss_history = []
acc_history = []

if TRAIN_ONESHOT_MODELS:
    # how long does this take?
    start_time = time.time()

    print("Training model")
    print("-" * 50)

    # for each training iteration...
    for i in range(1, n_iterations + 1):
        # make a training batch and train model using this batch
        inputs, targets = make_batch(train_images, batch_size)
        loss = model.train_on_batch(inputs, targets)

        # if it's time to evaluate the model...
        if i % interval == 0:
            # output some stats
            print("\n{}\n".format("-" * 20))
            print("Time for {} iterations: {:.2f} mins".format(i, (time.time() - start_time) / 60.))
            print("Train loss: {:.6f}".format(loss))

            # save model
            model.save_weights(os.path.join(model_dir, "weights.{}.h5".format(i)))
            
            # evaluate model
            val_acc = test_oneshot_model(model, n_classes, n_tasks, test_images, test_ranges)
            loss_history.append(loss)
            acc_history.append(val_acc)

            # update best accuracy
            if val_acc >= best_acc:
                print("Current best: {:.2f}; previous best: {:.2f}".format(val_acc, best_acc))
                best_acc = val_acc

    print("Elapsed time: {:.2f} minutes".format((time.time() - start_time) / 60.))
```

```python
if TRAIN_ONESHOT_MODELS:
    # show the training loss and accuracy over time
    graph_history(loss_history, acc_history, n_iterations, interval)
    
    # stats for best iteration
    best_iteration = interval * (np.argmin(loss_history) + 1)
    print("Best iteration :", best_iteration)
    print("Training loss  :", np.min(loss_history))
    print("Val. accuracy  :", acc_history[np.argmin(loss_history)])
    
    # load model weights by best loss
    model.load_weights(os.path.join(model_dir, "weights.{}.h5".format(interval * (np.argmin(loss_history) + 1))))

else:
    # show loss and acc histories
    display(Image(filename=os.path.join(os.getcwd(), "images", "history.png")))
    
    # load last model
    model.load_weights(os.path.join(model_dir, "weights.20000.h5"))
```

## Test nearest neighbor approach

```python
# return true if the nearest (euclidean) neighbor is the correct answer;
# otherwise, return false
def nearest_neighbor_correct(inputs, targets):
    # placeholder
    l2_distances = np.zeros_like(targets)
    
    # update l2_distances for each image pair
    for i in range(len(targets)):
        l2_distances[i] = np.sqrt(np.sum((inputs[0][i] - inputs[1][i]) ** 2))
    
    # true if the smallest l2_distance aligns with the actual target
    return np.argmin(l2_distances) == np.argmax(targets)
```

```python
# returns accuracy of nearest neighbor approach
def test_nn_accuracy(n, k):
    print("Evaluating NN approach on {} random {}-way one-shot learning tasks...".format(k,n))
    n_correct = 0
    
    # for each task...
    for i in range(k):
        inputs, targets = make_oneshot_task(n, test_images, test_ranges)
        n_correct += nearest_neighbor_correct(inputs, targets)
    
    nn_acc = 100. * n_correct / k
    print ("Nearest neighbor accuracy: {}%".format(nn_acc))
    return nn_acc
```

```python
train_accs, test_accs, nn_accs = [], [], []
tests = np.arange(1,21)
n_tasks = 50

# for each n-way one-shot learning task...
for n in tests:
    # compute and save accuracy values
    train_accs.append(test_oneshot_model(model, n, k=n_tasks, images=train_images, ranges=train_ranges))
    test_accs.append(test_oneshot_model(model, n, k=n_tasks, images=test_images, ranges=test_ranges))
    nn_accs.append(test_nn_accuracy(n, k=n_tasks))
    print("\n{}\n".format("-" * 20))

# also, compare to random guess approach
rand_accs = 100. / tests
```

```python
# create the figure
fig, ax = plt.subplots(figsize=(9,6))

# plot the results
ax.plot(tests, train_accs, color="c", label="Siamese (training set)", linestyle="-")
ax.plot(tests, test_accs, color="m", label="Siamese (testing set)", linestyle="--")
ax.plot(tests, nn_accs, color="y", label="Nearest neighbor", linestyle="-.")
ax.plot(tests, rand_accs, "k", label="Random guess", linestyle=":")

# pretty-fy the plot
ax.set_title("Accuracy vs. k in k-way one-shot learning")
ax.set_xlabel("k (number of possible classes)")
ax.set_ylabel("Accuracy (%)")
ax.set_xticks(tests)
ax.legend()

# let's see it
plt.savefig(os.path.join(os.getcwd(), "images", "results.png"), dpi=100)
plt.show()
```
