````python
# ---------------- Utils -------------------------------

#load data
def load_cifar10_data():
    """Loads the cifar10 image set and generates train and test data
    note that y is converted to categorical (dummy variables)
    :return: x_train, y_train, x_test, y_test, num_classes
    """
    num_classes = 10
    (x_train, y_train), (x_test, y_test) = cifar10.load_data()
    print('x_train shape:', x_train.shape)
    print(x_train.shape[0], 'train samples')
    print(x_test.shape[0], 'test samples')
    # Convert class vectors to binary class matrices.
    y_train = keras.utils.to_categorical(y_train, num_classes)
    y_test = keras.utils.to_categorical(y_test, num_classes)
    x_train = x_train.astype('float32')
    x_test = x_test.astype('float32')
    x_train /= 255
    x_test /= 255
    return x_train, y_train, x_test, y_test, num_classes


def show_cifar10_images(imageset = range(0, 9)):
    """
    Displays some of the images in CIFAR10
    :param imageset: list of 9 image indexes to display
    :return:
    """
    # load data
    (X_train, y_train), (X_test, y_test) = cifar10.load_data()
    print('X_train shape ', X_train.shape)
    # create a grid of 3x3 images
    for i in imageset:
        pyplot.subplot(330 + 1 + i)
        X = X_train.reshape(50000, 3, 32, 32).transpose(0,2,3,1).astype("uint8")
        pyplot.imshow(X_train[i].reshape(32, 32, 3), cmap=pyplot.get_cmap('gray'))
    # show the plot
    pyplot.show(block=False)


def plot_confusion_matrix(cm, class_labels, title='Confusion matrix',
                          cm_directory=os.path.join(os.getcwd(), 'model-performance'),
                          cm_filename='confusion_matrix.png', cmap=plt.cm.Blues):
    """
    Plot a confusion matrix
    :param cm: a confusion matrix generated by sklearn.metrics.confusion_matrix
    :param class_labels: set of text labels
    :param title: title of figure
    :param cmap: color map for the confusion matrix
    :return:
    """
    class_count = len(class_labels)
    fig = plt.figure(title)
    fig.set_size_inches(10, 8)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(class_count+1)
    plt.xticks(tick_marks, class_labels, rotation=45)
    plt.yticks(tick_marks, class_labels)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], horizontalalignment="center", color="white" if cm[i, j] > thresh else "black")
    #plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
    if not os.path.isdir(cm_directory):
        os.makedirs(cm_directory)

    plt.savefig(os.path.join(cm_directory, cm_filename), dpi=100)

    np.set_printoptions(precision=2)
    plt.show(block=False)


def save_CNN_model(model, save_dir=os.path.join(os.getcwd(), 'saved-models'),
                   model_filename='keras_cifar10_trained_model.h5'):
    """
    Save model and current weights
    :param model: Keras model
    :param save_dir: path name to save directory
    :param model_filename: filename for saved model
    :return: nothing
    """
    if not os.path.isdir(save_dir):
        os.makedirs(save_dir)
    model_path = os.path.join(save_dir, model_filename)
    model.save(model_path)
    print('Saved trained model at %s ' % model_path)


def load_CNN_model(save_dir, model_filename):
    # Load model and weights
    if not os.path.isdir(save_dir):
        os.makedirs(save_dir)
    model_path = os.path.join(save_dir, model_filename)
    model = load_model(model_path)
    print('Loaded trained model from %s ' % model_path)
    return model

# let's use this homework 1 function again
def graph_history(history, history_directory=os.path.join(os.getcwd(), 'model-performance'),
                  history_prefix="model_"):

    if not os.path.isdir(history_directory):
        os.makedirs(history_directory)
    
    # summarize history for accuracy
    fig = plt.figure()
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Tr. Acc', 'Val. Acc'])
    plt.savefig(os.path.join(history_directory, history_prefix + "accuracy_history.png"), dpi=100)
    plt.show()
 
    # summarize history for loss
    fig = plt.figure()
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Tr. Loss', 'Val. Loss'])
    plt.savefig(os.path.join(history_directory, history_prefix + "loss_history.png"), dpi=100)
    plt.show()

# write loss and accuracy to a file
def save_classification_metrics(metrics, 
                                metrics_directory=os.path.join(os.getcwd(), 'model-performance'),
                                metrics_filename="metrics.csv"):

    if not os.path.isdir(metrics_directory):
        os.makedirs(metrics_directory)

    metrics = pd.DataFrame(metrics, columns=["Value"], index=["Loss", "Accuracy"])
    metrics.to_csv(os.path.join(metrics_directory, metrics_filename))
````

````python
# ---------------- CNN Section -------------------------

def compile_shallow_CNN_model(input_shape, num_classes, batch_size=32):
    """Compiles a shallow CNN: conv - maxpool - conv - maxpool - dense - dense
    :param input_shape: tensor shape of the input
    (3 dim for 3 channel color images: colorChannel, height, width)
    :param num_classes: name of the h5 file containing the trained model params
    :param batch_size=32: size of the batches used for training
    :return: compiled model
    """
    model = Sequential()
    model.add(Conv2D(batch_size, kernel_size=(5, 5), strides=(1, 1),
                     activation='relu',
                     input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
    model.add(Conv2D(64, (5, 5), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dense(1000, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=keras.optimizers.SGD(lr=0.01),
                  metrics=['accuracy'])

    return model


def compile_deep_CNN_model(input_shape, num_classes, batch_size=32):
    """Compiles a simple deep CNN: with conv, maxpool, dropout
    :param input_shape: tensor shape of the input (not including observation count)
    (3 dim for 3 channel color images)
    :param num_classes: name of the h5 file containing the trained model params
    :param batch_size=32: size of the batches used for training
    :return: compiled model
    """
    model = Sequential()
    model.add(Conv2D(batch_size, (3, 3), padding='same',
                     input_shape=input_shape,
                     activation='relu'))
    # model.add(Activation('relu'))
    model.add(Conv2D(32, (3, 3), activation='relu'))
    # model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    # model.add(Activation('relu'))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    # model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.25))
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    # model.add(Activation('relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    # model.add(Activation('softmax'))
    # initiate RMSprop optimizer
    opt = keras.optimizers.rmsprop(lr=0.0001, decay=1e-6)
    # compile the model using RMSprop
    model.compile(loss='categorical_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])

    return model
````

To the training function, I added an early stopping callback which monitors validation loss and stops training if said loss does not improve in 25 epochs. I also split the nontest data into training and validation sets; this ensures we don't test our trained models on previously-seen data. 

Additionally, the training function now returns the results of the model-fitting process; this allows for better visualization of the results.

Furthermore, I defined a function ```save_classification_metrics()``` (defined in the Utilities cell above) which writes the validation loss and accuracy to a file. The evaluation function now calls this new function to save more information about the training process.

I also restructured the main code section. Specifically, I moved the code into a function called ```convnet()``` to which we pass an instantiated model, a filename, and a model name. Doing so allows us to call the function for each of the three models without changing filenames and function calls within the code cell. In other words, this restructure makes this notebook a little more user-friendly.

When the user passes a model to ```convnet()```, he or she can monitor the training/validation loss/accuracy for each epoch. When the model finishes training (either because it completes every epoch or because the early stopping callback triggers), the function will display history plots, classification metrics, and confusion matrices for the user. Additionally, the function saves these performance evaluations to various files. 

````python
def train_cifar10(model, x_train, y_train, x_val, y_val, batch_size=32, epochs=200, data_augmentation=False):
    """
    Train a simple deep CNN on the CIFAR10 small images dataset.
    :param model: Keras model to train
    :param x_train:
    :param y_train: categorical (dummy columns)
    :param x_val:  categorical (dummy columns)
    :param y_val:
    :param batch_size: =32
    :param data_augmentation: =False
    :return: trained model
    """
    
    history = None
    verbosity = 1

    callbacks = [EarlyStopping(
        monitor='val_loss',
        patience=25, 
        restore_best_weights=True
    )]

    if not data_augmentation:
        print('Not using data augmentation.')
        history = model.fit(x_train, y_train,
            batch_size=batch_size,
            epochs=epochs,
            validation_data=(x_val, y_val),
            shuffle=True,
            verbose=verbosity,
            callbacks=callbacks)

    else:
        print('Using real-time data augmentation.')
        # This will do preprocessing and realtime data augmentation:
        datagen = ImageDataGenerator(
            featurewise_center=False,  # set input mean to 0 over the dataset
            samplewise_center=False,  # set each sample mean to 0
            featurewise_std_normalization=False,  # divide inputs by std of the dataset
            samplewise_std_normalization=False,  # divide each input by its std
            zca_whitening=False,  # apply ZCA whitening
            rotation_range=0,  # randomly rotate images in the range (degrees, 0 to 180)
            width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
            height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
            horizontal_flip=True,  # randomly flip images
            vertical_flip=False)  # randomly flip images

        # Compute quantities required for feature-wise normalization
        # (std, mean, and principal components if ZCA whitening is applied).
        datagen.fit(x_train)

        # Fit the model on the batches generated by datagen.flow().
        history = model.fit_generator(datagen.flow(
            x_train, y_train,
            batch_size=batch_size),
            steps_per_epoch=x_train.shape[0] // batch_size,
            epochs=epochs,
            validation_data=(x_val, y_val),
            verbose=verbosity,
            callbacks=callbacks)

    return model, history


def eval_cifar10(model, x_test, y_test, cm_filename, metrics_filename):
    """Loads and evaluates a model on cifar10 dataset
    :param save_dir: subfolder the model is located in
    :param model_filename: name of the h5 file containing the trained model params
    :return: nothing is returned from this function
    """
    #initialization
    num_predictions = 20  # sample predictions

    print(x_test.shape[0], 'test samples')
    print("y_test shape: ", y_test.shape)

    # Load label names to use in prediction results
    label_list_path = 'datasets/cifar-10-batches-py/batches.meta'
    keras_dir = os.path.expanduser(os.path.join('~', '.keras'))
    datadir_base = os.path.expanduser(keras_dir)
    if not os.access(datadir_base, os.W_OK):
        datadir_base = os.path.join('/tmp', '.keras')
    label_list_path = os.path.join(datadir_base, label_list_path)
    with open(label_list_path, mode='rb') as f:
        labels = pickle.load(f)

    # report evaluation metrics
    evaluation = model.evaluate(x_test, y_test)
    print()
    print('------------------')
    print('Model accuracy = %.2f' % (evaluation[1]))
    print()
    print('Model metrics: ', evaluation)    # change this if you want other metrics

    # make predictions on all test inputs
    predictions = model.predict(x_test)
    print("Prediction shape: ", predictions.shape)

    # build class integers from target dummy variables
    actual_class_IDs = np.argmax(y_test, axis=1)
    # build class integers from predicted dummy variables
    predicted_class_IDs = np.argmax(predictions, axis=1)

    print("Actual class ID shape: ", actual_class_IDs.shape)
    print("Predicted class ID shape: ", predicted_class_IDs.shape)
    print("Labels names: ", labels['label_names'][0:10])

    # display label names
    lab_names = np.asarray(labels['label_names'][0:10])

    # generate & print confusion matrix to screen
    cm = confusion_matrix(actual_class_IDs, predicted_class_IDs)
    print(cm)

    # plot confusion matrix in pretty format
    plot_confusion_matrix(cm, class_labels=lab_names, title='Confusion matrix', cm_filename=cm_filename)
    
    # save classification metrics
    save_classification_metrics(metrics=evaluation, metrics_filename=metrics_filename)

    # ...... show some sample predictions (text only) .......
    for predict_index, predicted_y in enumerate(predictions):
        actual_label = labels['label_names'][np.argmax(y_test[predict_index])]
        predicted_label = labels['label_names'][np.argmax(predicted_y)]
        print('Actual Label = %s vs. Predicted Label = %s' % (actual_label,
                                                              predicted_label))
        if predict_index == num_predictions:
            break

    return
````

````python
# these are model-independent

#select activities to perform below
SHOW_SAMPLES = True  # show examples of CIFAR10 images
BUILD_MODEL = True  # overwrites current model with initialized parameters
TRAIN_MODEL = True  # overwrites current model by training
SAVE_MODEL = True   # saves current model to "model_filename". should only be used if a model is built or trained
LOAD_MODEL = False    # overwrites current model with loaded model
EVAL_MODEL = True    # evaluates performance of model on CIFAR10

# ------ initialization ------
seed = 10
np.random.seed(seed)
save_dir = os.path.join(os.getcwd(), 'saved-models')  # subdirectory for saved models

# ------ load data ------
x_nontest, y_nontest, x_test, y_test, num_classes = load_cifar10_data()
input_shape = x_nontest.shape[1:]
batch_size = 64

# --------- show sample images ----------------
if SHOW_SAMPLES:
    show_cifar10_images()
````

````python
# =========  MAIN RUNNING CODE  ==================

def convnet(model, model_filename, model_name):
    print("Starting script.")
    prefix = "keras_cifar10_" + model_name + "_"

    # ------------ Execute Training on compiled model -----------------------
    if TRAIN_MODEL:
        epochs = 200
        data_augmentation = False

        model, history = train_cifar10(model, x_train, y_train, x_val, y_val, batch_size, epochs, data_augmentation)
        history_filename = prefix + "training_performance.png"

        graph_history(history=history, history_prefix=prefix)
        print("Model Training Complete")

        # ------------- save the trained model ---------------------
        if SAVE_MODEL:
            save_CNN_model(model, save_dir, model_filename)

    # ---------- load a saved model later for evaluation --------
    if LOAD_MODEL:
        model_load_filename = prefix + "trained_model.h5"
        model = load_CNN_model(save_dir, model_load_filename)

    # ---------- evaluate the trained model on the test data ----------
    if EVAL_MODEL and model is not None:
        cm_filename = prefix + "confusion_matrix.png"
        metrics_filename = prefix + "classification_metrics.csv"
        eval_cifar10(model, x_test, y_test, cm_filename=cm_filename, metrics_filename=metrics_filename)

    # at end of code, this call to show keeps images on the screen
    plt.show(block=True)

    print("Done. Exiting...")
````

Note that all performance diagrams (that is, history plots, classification metrics, and confusion matrices for all three models) are displayed at the bottom of this notebook.

For both of the provided models and for my own model, I decided not to utilize data augmentation. Additionally, I used an early stopping callback for all three models. As mentioned previously, this callback monitors validation loss (categorical cross-entropy in our case); if the model's loss does not improve within 25 epochs, training terminates.

This is greatly preferred to simply training on 200 epochs per model. The shallow, deep, and student models stopped in 51, 129, and 52 epochs, respectively. In total, then, the early stopping callback eliminated 368 epochs. Because the 332 epochs that *did* run took over an hour, it's clear that all 600 epochs would've taken several hours. I improved my model over multiple training sessions, so the early stopping callback saved significant time.

Besides, letting ```patience = 20``` effectively guarantees that the loss returned is either a) very close to the best possible loss or b) the best possible loss itself. The history plots (shown below) illustrate this: given more epochs, it's not likely that any of the models would improve significantly because the loss/accuracy values either reach an asymptote or start to worsen.

Finally, the callback sets ```restore_best_weights = True``` to ensure that, even if the model worsens (i.e., begins to overfit), the *saved* model uses the weights of the epoch with the lowest loss (that is, it uses the weights from the model as it was 10 epochs before training terminated).
