```python
from random import randint
import matplotlib.pyplot as plt
from statistics import mean, variance, stdev
%matplotlib inline
```

```python
counts = []

for i in range(1, 7):
    for j in range(1, 7):
        for k in range(1, 7):
            counts.append(i + j + k)

m = mean(counts)
v = variance(counts)
s = stdev(counts)

print("mean:", m)
print("variance:", v)
print("std variance:", s)

plt.hist(counts, 16, color='c')
plt.title("3d6 sums")
plt.xlabel("sum")
plt.ylabel("frequency")

for i in range(-3, 4):
    plt.axvline(m + s * i, linestyle='--')
```

```python
sims = []
for i in range(1000):
    sims.append([])
    for j in range(10):
        sims[i].append(randint(1,7) + randint(1,7) + randint(1,7))
    
means = []
variances = []

for i in sims:
    means.append(mean(i))
    variances.append(variance(i))

big_mean = mean(means)
big_stdev = stdev(means)
big_variance = big_stdev ** 2

plt.hist(means, color='c')
plt.title("Means")
plt.axvline(big_mean, color='r')
plt.axvline(m, color='b')
plt.show()

plt.hist(variances, color='c')
plt.title("Variances")
plt.axvline(big_variance, color='r')
plt.axvline(v, color='b')
plt.show()
```
