```python
# STEP 0  - Load data & convert categorical features - instructor provided code

state = 1  #random seed
Hitters = pd.read_csv('ISLR_Hitters.csv',index_col=0)
Hitters = Hitters.dropna(how='any')
Hitters = Hitters.reset_index()

Hitters = Hitters.drop(['index'],axis=1)
Hitters['League'] = Hitters['League'].map({'A': 0, 'N': 1})
Hitters['Division'] = Hitters['Division'].map({'E': 0, 'W': 1})
Hitters = Hitters.drop(['NewLeague'],axis=1)
```

```python
# helper functions for transformation
# the transformed salary should work better for model fitting
def salary_transform(salary):
    return np.log10(salary)

# the untransformed salary is needed for test set performance reporting and residuals analysis
def salary_untransform(transformed_salary):
    return 10 ** transformed_salary

# transform the salary
Hitters["Salary"] = salary_transform(Hitters["Salary"])
```

# Train the model & Tune Hyperparameters using Cross-Validation

## STEP 5.
Using sklearn k-fold split (sklearn.model_selection.KFold), write code to set up a k-fold cross-validation with the goal of choosing the best hyperparameters for a random forest model (sklearn.ensemble.RandomForestRegressor).

Select and provide rationale for your choice for n_splits based on amount of data you have available in the non-test set.  Your goal is to determine the best combination of two parameters: maximum tree depth (max_depth), and the number of features to consider at each split (max_features). 

The hyperparameter  max_depth of the trees should include integer values from 1 to 20, and your exploration over max_features should include values from 1 to p (all features). You can decide whether to fix the number of trees (n_estimators) or include it as a third hyperparameter to explore (it should start with at least 100 but you may want to explore higher values if you will tune this hyperparameter with cross-validation) – then explain whether you are selecting a specific value or tuning this value with cross validation.

Since you will use a cross-validation wrapper to tune hyperparameters, set oob_score to False in the initialization call to RandomForestRegressor. For each tuple of (max_depth, max_features), compute and collect the mean k-fold cross-validation MSE using predict().  

```python
# first, split into X and Y sets
train_x = Hitters_nonTest.drop(columns=["Salary"])
train_y = Hitters_nonTest["Salary"]

test_x = Hitters_test.drop(columns=["Salary"])
test_y = Hitters_test["Salary"]
```

```python
# capture the values
depths = []
features = []
mses = []

# let's see how long this takes (just because)
start = timer()

# for every depth...
for d in range(1, 21):
    print("depth:", d, end=", ")
    print("features:", end=" ")

    # for every number of features...
    for f in range(1, 19):
        print(f, end=" ")

        # initialize 10-fold CV
        kf = KFold(n_splits=10, shuffle=False,  random_state=state)
        kf.get_n_splits(train_x)

        # initialize random forest regression
        reg = RFR(max_depth=d, max_features=f, n_estimators=100, oob_score=False, random_state=state)
        kfold_mses = []

        # for every fold...
        for train_idx, test_idx in kf.split(train_x):
            # fit the regression model
            reg.fit(train_x.iloc[train_idx], train_y.iloc[train_idx])

            # predict salary and compute the MSE
            kfold_mses.append(MSE(train_y.iloc[test_idx], reg.predict(train_x.iloc[test_idx])))

        # save these for later
        depths.append(d)
        features.append(f)
        mses.append(np.mean(kfold_mses))

    print()

# how long did it take?
print("\n", timer() - start, " seconds", sep="")
```

#### Select and provide rationale for your choice for $n\_splits$ based on the amount of data you have available in the non-test set.

When $k=10$, $k$-fold cross-validation trains on 90% of the training set and validates on the remaining 10%. Because are 200 observations in the non-test set, each of the ten validation sets contain $200\times0.10=20$ observations. Twenty observations is not so few as to unnecessarily increase variance, and using 10-fold CV means that we limit our overall execution time.

For these reasons, I think $k=10$ provides a good balance.

#### Explain whether you are selecting a specific value for the number of trees ($n\_estimators$) or tuning this value with cross-validation.

I use $n\_estimators=100$. I did execute the above code within a loop over $n\_estimators$ values of 100, 150, 200, 250, and 300, but this required 46 minutes to execute. The minimum MSE found was 0.03693448017377648. As we see in step 7, the MSE decreases by a negligible amount; the small decrease is not worth the ridiculous increase in execution time.

For this reason, using the specific value $n\_estimators=100$ is sufficient.

```python
# contour plot
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)

# pretty-fy the plot
fig.suptitle("Contour Plot of $max\_features$ vs. $max\_depth$ vs. Mean Squared Error")
fig.subplots_adjust(top=0.9)

ax.set_xlabel("$max\_features$")
ax.set_ylabel("$max\_depth$")

ax.set_xticks(np.arange(2,19,2))
ax.set_yticks(np.arange(2,21,2))

# plot it
contour = ax.contourf(np.arange(1,19,1), np.arange(1,21,1), np.array(mses).reshape((20,18)),
    cmap=cm.RdYlGn_r, levels=100)
plt.colorbar(contour, ax=ax)

# show the minimum MSE
ax.plot(features[np.argmin(mses)], depths[np.argmin(mses)], color="w", markersize=10, marker="x")
ax.annotate("Minimum MSE", xy=(features[np.argmin(mses)], depths[np.argmin(mses)]),
    xytext=(features[np.argmin(mses)], depths[np.argmin(mses)] + 0.5), color="w")
```

Using code, determine and report the quality of the model for predicting salary on the sequestered test set.  Don’t forget to handle the log transformation you did in data preprocessing – your performance values should be based on real dollars (not log-transformed dollars).

## STEP 10.
Develop a scatterplot of the regression residuals:  The figure’s x axis expresses the true dollar amount of salary, the figure’s y axis represents the prediction error (positive values mean overprediction, negative values mean underprediction, and $y=0$ would mean correct prediction). 

```python
# create and pretty-fy the plot
fig, ax = plt.subplots(1, 1, figsize=(9,6))
ax.set_title("Residual Errors in Predicting Player Salary", fontsize=20)
ax.set_xlabel("Actual Salary (in thousands of dollars)")
ax.set_ylabel("Residual Error")

# edges of the plot
min_x = -0.1 * y_actual.min()
max_x = 1.1 * y_actual.max()
min_y = 1.1 * y_hat.min()
max_y = 1.1 * y_hat.max()

# bound the plot
ax.axis([min_x, max_x, min_y, max_y])

# plot the horizontal and vertical lines
ax.plot([min_x, max_x], [0, 0], color='k', linestyle="--")
ax.plot([700, 700], [min_y, max_y], color='k', linestyle="--")

# plot the salary errors
ax.scatter(y_actual, y_hat, alpha=0.75, edgecolors='m', facecolors='w')
```
